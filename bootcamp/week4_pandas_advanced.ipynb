```python
import pandas as pd

customers = pd.DataFrame({
    "customer_id": [1, 2, 3, 4, 5],
    "customer_name": ["Alice", "Bob", "Carol", "Dave", "Eve"],
    "region": ["North", "South", "East", "West", "North"]
})

orders = pd.DataFrame({
    "order_id": [101, 102, 103, 104, 105, 106],
    "customer_id": [1, 2, 1, 3, 5, 6],
    "product": ["T-Shirt", "Jeans", "Sneakers", "Jacket", "Hat", "Boots"],
    "revenue": [59.97, 99.98, 74.99, 44.43, 29.99, 89.99]
})
```

### Exercise 1

```python
print(customers)

print(orders)
```

### Exercise 2

```python
print(pd.merge(customers, orders, on="customer_id", how= "inner"))
```

```python
print(pd.merge(customers, orders, on="customer_id", how= "left"))
```

```python
print(pd.merge(customers, orders, on="customer_id", how= "right"))
```

```python
print(pd.merge(customers, orders, on="customer_id", how= "outer"))
```

### Exercise 3

```python
df_merged = pd.merge(customers, orders, on="customer_id", how= "inner")
df_merged
```

```python
pivot = df_merged.pivot_table(
    values = "revenue",
    index = "customer_name",
    columns = "region",
    aggfunc = "sum",
    fill_value = 0
)
print(pivot)
```

### Exercise 4

```python
def revenue_tier(revenue):
    if revenue >= 100:
        return "High"
    elif revenue >= 50:
        return "Medium"
    else:
        return "Low"

df_merged["revenue_tier"] = df_merged["revenue"].apply(revenue_tier)

print(df_merged[["customer_name", "product", "revenue", "revenue_tier"]])
```

### Exercise 5

```python
df_merged["running_revenue"] = df_merged["revenue"].cumsum()
df_merged["revenue_rank"] = df_merged["revenue"].rank(ascending=False)
df_merged["rolling_avg"] = df_merged["revenue"].rolling(window=2).mean()

print(df_merged[["customer_name", "product", "revenue", "running_revenue", "revenue_rank", "rolling_avg"]])
```

### Exercise 6

```python
df_merged.to_csv("order_analysis.csv", index=False)

df_check = pd.read_csv("order_analysis.csv")

print(df_check)
```

### Practice Project

```python
def final_revenue(row):
  if row["revenue_tier"] == "High" or row["revenue_tier"] == "Medium":
    return row["revenue"] * 1.10
  else:
    return row["revenue"]

df_merged["final_revenue"] = df_merged.apply(final_revenue, axis=1)

print(df_merged)
```

```python
print(df_merged.groupby("region")["final_revenue"].sum())
```

```python
pivot = df_merged.pivot_table(
    values = "revenue",
    index = "customer_name",
    columns = "revenue_tier",
    aggfunc = "sum",
    fill_value = 0
)
print(pivot)
```
