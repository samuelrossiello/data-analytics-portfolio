### Practice Project

```python
products = ["T-Shirt", "Jeans", "Jacket", "Sneakers", "Hat"]
prices = [19.99, 49.99, 89.99, 74.99, 14.99]
```

```python
print(len(products))
```

```python
for product, price in zip(products, prices):
  print(f"{product}: ${price}")
```

```python
def price_category(price):
  if price > 75:
    return "Premium"
  elif price < 25:
    return "Budget"
  else:
    return "Mid-range"

print(price_category(50))
```

```python
for product, price in zip(products, prices):
  category = price_category(price)
  print(f"{product}: ${price} - {category}")
```
