### Exercise 2

```python
import pandas as pd

data = {
    "product": ["T-Shirt", "Jeans", "Jacket", "Sneakers", "Hat"],
    "price": [19.99, 49.99, 89.99, 74.99, 14.99],
    "category": ["Clothes", "Clothes", "Clothes", "Footwear", "Accessories"],
    "units_sold": [150, 85, 40, 120, 200]
}

df = pd.DataFrame(data)

df
```
