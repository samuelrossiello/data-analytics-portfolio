### Introduction

```python
import pandas as pd

data = {
    "order_id": [1001, 1002, 1003, 1004, 1005, 1006, 1007, 1008, 1009, 1010],
    "customer": ["Alice", "Bob", "Carol", "Alice", "Dave", "Bob", "Eve", "Carol", "Dave", "Alice"],
    "category": ["Clothes", "Footwear", "Clothes", "Accessories", "Footwear", "Clothes", "Accessories", "Footwear", "Clothes", "Accessories"],
    "product": ["T-Shirt", "Sneakers", "Jeans", "Hat", "Boots", "Jacket", "Belt", "Sandals", "Chinos", "Scarf"],
    "price": [19.99, 74.99, 49.99, 14.99, 89.99, 89.99, 24.99, 39.99, 54.99, 29.99],
    "units": [3, 1, 2, 5, 1, 1, 2, 3, 2, 4],
    "discount": [0.0, 0.1, 0.0, 0.05, 0.2, 0.0, 0.0, 0.15, 0.1, 0.0],
    "region": ["North", "South", "East", "North", "West", "East", "South", "North", "West", "East"]
}

df = pd.DataFrame(data)
df.head(10)
```

### Exercise 2

```python
import numpy as np

df.loc[2, "discount"] = np.nan
df.loc[5, "price"] = np.nan
df.loc[8, "region"] = np.nan
```

```python
print(df.isnull().sum())
```

### Exercise 3

```python
df["discount"] = df["discount"].fillna(0.0)
print(df)
```

```python
df["price"] = df["price"].fillna(df["price"].mean())
print(df)
```

```python
df = df.dropna()
print(df)
```

```python
print(df.isnull().sum())
```

### Exercise 7

```python
df["revenue"] = df["price"] * df["units_sold"]
```

```python
print(
    df.groupby("category").agg(
    total_revenue = ("revenue", "sum"),
    avg_price = ("price", "mean"),
    total_units = ("units_sold", "sum"),
    order_count = ("product", "count")
  )
)
```
